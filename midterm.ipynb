## 2)

```python
n_sky,sky_bins,patches = p.hist(sky.flatten(),bins=60,normed=True)
p.xlabel('Sky Pixel Values')
p.ylabel('Relative Frequency')
```

```python
#define a gaussian pdf function
def gaussianPDF(mu, sigma, x):
    z=(x - mu)/sigma
    return 1./(sigma * n.sqrt(2.0*n.pi))*n.exp(-z**2/2.0)
#define the maximum likelihood function that has two parameters to fit (mu,sigma)
def mll(theta, data):
    mu = theta[0]
    sigma = theta[1]
    pxs = n.log(gaussianPDF(mu, sigma, data))
    return -n.sum((pxs))
#make an initial guess of the mean and sigma from the histogram
initial = [1225.0,5.0]
hb=opt.minimize(mll,initial,args=(sky.flatten(),),method='Nelder-Mead')

print 'Fitted mu:', hb.x[0]
print 'Fitted sigma:', hb.x[1]
```

## 3)

```python
#try the Anderson-Darling non-parametric test with a significance level of 5%
A2, sig, crit_val = stats.anderson(sky_bins,dist='norm')
print A2
print sig
print crit_val
```

From the Anderson-Darling test, the A2 statistic is larger than the 5% significance level so the hypothesis of an underlying normal distribution can be rejected at the 0.05 significance level.

```python
#try the Shapiro-Wilk test with a significance level of 5%
W,p_sw = stats.shapiro(sky_bins)
print W
print p_sw
```

From the Shapiro-Wilk test, the p_value is less than 0.05 so we can reject the null hypothesis that the distribution of the sky pixels is from a Gaussian distribution at the 5% level.

## 4)

```python
#assume the sky background is Gaussian with the parameters choosen from the ML estimate in part 2) (mu = 1230, sigma =7.31)
back_dist = stats.norm(1230.0, 7.28)
p_bh = 1.0 - back_dist.cdf(im.flatten())
p_bh_sorted = n.sort(p_bh)
# plot the cutoffs for epsilon = 0.01
ax = p.axes(xscale='log', yscale='log')
p_reg_over_eps = 10 ** n.linspace(-3, 0, 10000)
#contamination (epsilon) is 1%
epsilon = 0.01
x = p_reg_over_eps * epsilon
y = p_reg_over_eps
ax.plot(x, y, '--k',label='$\epsilon = 0.01$')
cp = n.linspace(0, 1, n.size(p_bh_sorted))
ax.plot(p_bh_sorted, cp, '-g',label='Background + Sources')
ax.plot(p_bh_sorted, p_bh_sorted, ':k',label='Background')
ax.set_xlim(1E-7, 1)
ax.set_ylim(1E-3, 1)
ax.set_xlabel('$p = 1 - H_B(i)$')
ax.set_ylabel('normalized C(p)')
p.title('Benjamini and Hochberg Method')

#we want to find the intersection of the Background+Sources and contamination line
#interpolate both of the functions
f1=interp.interp1d(p_bh_sorted,cp)
f2=interp.interp1d(x,y)
#define a function that finds the intersection
def findIntersection(fun1,fun2,x0):
 return fsolve(lambda x : fun1(x) - fun2(x),x0)

p_crit = findIntersection(f1,f2,.00003)
#plot the critical p-value on the graph
p.plot(p_crit, p_crit/epsilon,'ro',label="Critical p-value")
p.legend(loc=2)
```

```python
print "Critial P-value:", p_crit
```

Selecting all the pixels with p values lower than $5.6\times10^{-5}$ will result in a sample that has no more than 1% contamination.

```python
uncont = n.where(p_bh <= p_crit)
mask = n.copy(im.flatten())*0.0
mask[uncont] = 1.0
#select only these pixels
good_pix = im.flatten()*mask
#reshape the good_pixels
good_pix_im = n.reshape(good_pix,n.shape(im))
#show the good pixels scalled so you can see them
f, axs = p.subplots(2,2,figsize=(15,15))
p.subplot(2,1,1)
p.imshow(good_pix_im**.2/n.max(good_pix_im**.2))
p.title("Masked Uncontaminated Image (scalled)")
p.colorbar()
p.summer()
p.subplot(2,1,2)
p.imshow(n.log2(im))
p.title("Original Image (scalled)")
p.colorbar()
p.spectral()
```

By comparing the two images, we see that we are indeed selecting what appears to be only sources.

```python
#only use a subsample of the galaxies (Mr_sub and z_sub are the galaxies below the detection limit)
cut = where((z_sub > 0.017) & (z_sub < 0.0185) & (Mr_sub < -16.8))
z_schec = z_sub[cut]
Mr_schec = Mr_sub[cut]
#fit a Schechter profile to this distribution, use MLE to find the M* and alpha
#define the Schechter function in absolute magnitudes
#we need to normalie the function by dividing by int[Phi,{Lmin,inf}] (i.e. find phi*)
def schecPDF(Mstar, alpha, M):
    MstarM = 0.4 * (Mstar - M)
    c1 = 0.4*n.log(10.)
    c2 = 10.**(MstarM*(alpha+1.))
    c3 = n.exp(-10.**MstarM)
    norm = n.float(mpmath.gammainc(alpha+1,10**(0.4*(Mstar+16.8))))
    return c1*c2*c3/norm
#define the maximum likelihood function that has two parameters to fit (Mstar,alpha)
#phistar is a normalization constant
def mll(theta, data):
    Mstar = theta[0]
    alpha = theta[1]
    pxs = n.log(schecPDF(Mstar, alpha, data))
    return -n.sum((pxs))
#make an initial guess of the mean and sigma from the histogram
initial = [-20., -1.0]
bnds = [(-24.,-19.),(-2.,0.)]
schec_params = opt.minimize(mll,initial,args=(Mr_schec,), method="L-BFGS-B")
M_star_fit = schec_params.x[0]
alpha_fit = schec_params.x[1]
print 'Convergence:', schec_params.success
print 'Fitted M_star:', M_star_fit
print 'Fitted alpha:', alpha_fit
```

The best fit for $M^*$ is -21.0 and the best fit for $\alpha$ is -1.15

## 6)

Use the graphical method to estimate the errors on $M^*$ and $\alpha$

```python
#Setup a grid for computing the likelihood
extent = [-22,-20,-1.0,-1.5]
Mstar_range,alpha_range = n.mgrid[extent[0]:extent[1]:250j,extent[2]:extent[3]:250j]
like_range = Mstar_range*0
# Evalute the likelihood for each point on the grid
for i in range(like_range.shape[0]):
    for j in range(like_range.shape[1]):
        like_range[i,j] = -1*mll([Mstar_range[i,j],alpha_range[i,j]], Mr_schec)
        
# Find the maximum
max_like = n.max(like_range)
[Mstar_best] = Mstar_range[like_range == max_like]
[alpha_best] = alpha_range[like_range == max_like]
# Print best fit
print "Best fit parameters from brute force:",Mstar_best, alpha_best
```

The brute force method agrees with the MLE method.

```python
# plot results
im_grid = p.imshow(like_range.T,cmap=p.cm.jet,
                vmin=n.max(like_range)*1.25,vmax=n.max(like_range),
                extent=extent,origin='lower')
p.scatter(Mstar_best,alpha_best,s=50,c='k',marker='+',label='Best Fit')
cbax = p.colorbar(im_grid,orientation='horizontal',fraction=0.05,aspect=35,pad=0.15)
cbax.set_label('Log Likelihood')
p.xlabel('Mstar')
p.ylabel('alpha')
p.xlim(extent[0],extent[1])
p.ylim(extent[2],extent[3])
p.gcf().set_size_inches(10,5)
p.contour(Mstar_range,alpha_range,like_range,levels=[n.max(like_range)-.5],linestyles='solid',colors='black')
#find 1 sigma error, where log Likelihood is equal to the maximum - 0.5 
inner = where(like_range > n.max(like_range)-.5)
#Mstar +/- 1sigma values 
print "sigma Mstar = ", Mstar_range[max(inner[0]),0] - Mstar_range[min(inner[0]),0]
#alpha +/- 1sigma values
print "sigma alpha = ", alpha_range[0,min(inner[1])] - alpha_range[0,max(inner[1])]

p.scatter(Mstar_range[min(inner[0]),0],alpha_range[0,max(inner[1])],s=50,c='k',marker='*',label="+ 1sigma")
p.scatter(Mstar_range[max(inner[0]),0],alpha_range[0,min(inner[1])],s=50,c='k',marker='*',label="- 1 sigma")
p.legend()
```

The plot above shows the log Likelihood as a funtion of the Schechter parameters.  The black contour is the $1\sigma$ error,  $\sigma_{M^*} = 0.265$ and $\sigma_{\alpha} = 0.08$.  As seen from the above plot, $\alpha$ and $M^*$ are clearly correlated.

## 7)

If we were to include all of the galaxies in our sample between $0.017 < z < 0.018$ we could still use the MLE algorithm but we would have to account for the fact that we would be biased towards selecting bright galaxies over faint galaxies.  Therefore, we should correct for this by weighing the bright galaxies less than the faint galaxies.  One way to do this is using the $V_{max}$ method.  Since galaxies with brighter magnitudes can be seen to larger distances, we shold weigh each galaxy's  contribution to the luminosity function by 1/Vmax, where Vmax is the maximum volume over which the objects could have been seen. Brighter objects will have a larger volume and therefore will be given less weight through this method since these bright objects will be more fully sampled.  $V_{max}$ will be a sphere with a radius found from the distance modulus, using the object’s absolute magnitude and the limiting apparent magnitude.

```python
#use chi squared minimization to fit the Schechter function to the binned histogram
schec_hist = n.histogram(Mr_schec,bins=20,normed=True)
#find the bin edges and centers
schec_bin_edges = schec_hist[1]
schec_bin_centers = (schec_bin_edges[:-1] + schec_bin_edges[1:])/2.
#define the Schechter function
schec_func = lambda x,phiStar,MStar,alpha: 0.4*n.log(10.)*phiStar*10.**((.4*(MStar-x))*(alpha+1.))*n.exp(-10.**(.4*(MStar-x)))
#p0 = [phi*,M*,alpha]
#We want to minimize chi squared so we can use curve_fit which implements a non-linear least squares fit
popt,pcov = opt.curve_fit(schec_func,schec_bin_centers,schec_hist[0],p0=[1.0,-20.0,-1.0])
print "Optimized M*:", popt[1]
print "Optimized alpha:", popt[2]
p.plot(schec_bin_centers,n.log10(schec_hist[0]),'r.',label='Data')
x = n.linspace(-23,-16,100)
phi = n.copy(x)
phi = 0.4*n.log(10.)*popt[0]*10.**((.4*(popt[1]-x))*(popt[2]+1.))*n.exp(-10.**(.4*(popt[1]-x)))
p.plot(x,n.log10(phi),'k-',label='Reduced chi^2 Schechter Function')
p.xlabel('Absolute Magnitude')
p.ylabel('Log Relative Frequency')
p.title('$\chi^2$ Method')
p.legend(loc=4)
```

Compared to the best-fit paramters obtained from the MLE method, the $\chi^2$ method returns a lower (brighter) $M^*$ and a lower $\alpha$ value.  $M^*$ is about 0.5 mag. brighter ($2\%$) and $\alpha$ is about 0.08 lower ($6.5\%$) when using the $\chi^2$ method.  The $M^*$ value is not within the errors, however, the $\alpha$ value is within the $1\sigma$ error.
