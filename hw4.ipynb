```python
#read in the file columns are Mass, Mass_err, Radius, Radius_err
data = ascii.read('table_1.dat')
M = data['Mass']
M_err = data['Mass_err']
R = data['Radius']
R_err = data['Radius_err']
mu = n.mean(R)
#Assume the stellar radii follow a Gaussian distribution
#we are after the mean radius, and will assume the standard deviation is fixed
sigma_R = n.std(R)
#define the pdf of the radius as a Guassian
def radiusPDF(theta_0, theta_1, sigma, M, R):
    mu = theta_0 + theta_1*M
    pp  = 1./(sigma * n.sqrt(2 * n.pi)) * n.exp( - (R - mu)**2 / (2 * sigma**2) )
    return pp

#define the Log(likelihood) so to include all the data points the Log(likelihood) becomes a sum
def logl(theta_0, theta_1, sigma, M, R):
    pxs = n.log(radiusPDF(theta_0, theta_1, sigma, M, R))
    return n.sum((pxs))

#Setup a grid for computing the posterior
extent = [-2.0,2.0,-2.0,2.0]
#assume the prior probabilities of theta_0 and theta_1 are uniform
p_theta0 = 1.0/(extent[1]-extent[0])
p_theta1 = 1.0/(extent[3]-extent[2])
theta0_range,theta1_range = n.mgrid[extent[0]:extent[1]:250j,extent[2]:extent[3]:250j]
posterior = theta0_range*0
# Evalute the Log(likelihood) for each point on the grid
for i in range(posterior.shape[0]):
    for j in range(posterior.shape[1]):
        #include the prior probabilities in the posterior, which are just constants
        posterior[i,j] = logl(theta0_range[i,j],theta1_range[i,j], sigma_R, M, R) + n.log(p_theta0) + n.log(p_theta1)
#convert the log posterior back to the posterior
posterior = n.e**posterior
```

```python
#normalize the posterior distribution
normC = 1.0/(integ.simps(integ.simps(posterior,theta1_range[0,:],axis=1),theta0_range[:,0]))
posterior *= normC
#define the marginal pdfs
p0 = integ.simps(posterior,theta1_range[0,:],axis=1)
p1 = integ.simps(posterior,theta0_range[:,0],axis=0)
cdf0 = integ.cumtrapz(p0,theta1_range[0,:])
cdf1 = integ.cumtrapz(p1,theta0_range[:,0])
h0 = theta0_range[n.min(n.where(cdf0>=0.5)),0]
sig_h0p = theta0_range[n.min(n.where(cdf0>=0.683)),0]
sig_h0p2 = theta0_range[n.min(n.where(cdf0>=0.955)),0]
sig_h0m = theta0_range[n.min(n.where(cdf0>=0.317)),0]
sig_h0m2 = theta0_range[n.min(n.where(cdf0>=0.045)),0]
h1 = theta1_range[0,n.min(n.where(cdf1>=0.5))]
sig_h1p = theta1_range[0,n.min(n.where(cdf1>=0.683))]
sig_h1p2 = theta1_range[0,n.min(n.where(cdf1>=0.955))]
sig_h1m = theta1_range[0,n.min(n.where(cdf1>=0.317))]
sig_h1m2 = theta1_range[0,n.min(n.where(cdf1>=0.045))]
```

# 2. 1)

This approach is not accounting for error propagation in the measurement of f.  f depends on the number of binary star systems and the total number of stars in the sample.  To correctly estimate the error on the measurement of f one should use the propagation of errors equation, where the error on f will be:

$\sigma_f = \sqrt{(\frac{\partial f}{\partial B})^2 \sigma_B^2 + (\frac{\partial f}{\partial N})^2 \sigma_N^2}$

Where $B$ is the number of binary stars measured, $N$ is the total number of stars in the sample, and $f = \frac{B}{N}$.

If where are assuming Poisson statistics then $\sigma_x$ is equal to the square-root of the number observed and the error on f can be written as:

$\sigma_f = \sqrt{\frac{B}{N^2} + (\frac{B}{N})^2}$

Which in this case gives: $\sigma_f = 0.165$ rather than $0.378$

## 2)

Bayes' theorem for this problem whould be:

$\Large P(~f\mid \{x_i\}) = \frac{P(\{x_i\}\mid ~f) P(f)}{P(\{x_i\})}$

Where the likelihood, $P(\{x_i\}\mid f)$, is a binomial distribution, and it is best to assume the prior probability on the binary fraction of stars, P(f), is uniform between $0-1$.  $P(\{x_i\})$ is just a normalization term.

Then the posterior probability can be written as:

$\large P(~f\mid \{x_i\}) \propto f^B(1-f)^{N-B}$

The proportional sign indicates that the posterior will need to be properly normalized.

```python
def like(B,N,f):
    binom = special.binom(N,B)
    hh = f**B*(1.-f)**(N-B)
    pp = binom * hh
    pp = pp/pp.max()
    return pp

def prior_uniform(bins):
    return n.ones(bins)

def bayes(B, N):
    """bayes(): bayesian coin toss example"""
    # Number of bins over which we calculate the posterior pdf
    bins = 1001
    n.random.seed()
    prior = prior_uniform(bins)
    # Create an array f of bin bins over which we calculate the pdf, each of which is 1/bin wide 
    f = n.linspace(0,1,bins,endpoint=True)
    # Calculate likelihood across all bins
    ll = like(B, N, f)
    # Now calculate posterior pdf from likelihood and prior
    post = ll * prior
    
    # Normalize the posterior
    normalized = integ.simps(post,f)
    post /= normalized

    # calculate the mean, uncertainty
    cumulative = n.array([integ.simps(post[:i],f[:i]) for i in range(1,post.size)])
    f0 = f[n.min(n.where(cumulative>=0.5))]
    sig_minus = f[n.min(n.where(cumulative>=0.16))]
    sig_plus = f[n.min(n.where(cumulative>=0.68))]
    sig = (sig_plus-sig_minus)/2.0
    
    #f0 = float(B)/float(N) 
    #sig = n.sqrt(float(B)/float(N)**2 + (float(B)/float(N)**2))
    
    # Plot it
    p.plot(f, prior, c='g', lw=2.0, label='Prior')
    p.plot(f, post, c='r', lw=2.0, label='Posterior')
    p.axhspan(0, n.max(post)+.5, sig_minus, sig_plus, facecolor='0.7', alpha=0.5, label=r'$\bar{\sigma}$ = %.3f' % sig)
    p.vlines(f0,0,n.max(post)+.5,linestyles='dashed',label=r'$f_0$ = %.3f' % f0)
    p.ylim(0,n.max(post)+.5)
    #p.xlabel('')
    #p.ylabel('prob(H|{data},I)')
    #p.title('Bayes example: coin toss -- N = %i' % N)
    p.legend()
    p.gcf().set_size_inches(12,6)
    p.show()
    print "The best estimate for f is:", f0
    print "With and +$\sigma$ error of:", sig_plus-f0
    print "With and -$\sigma$ error of:", f0-sig_minus
```
